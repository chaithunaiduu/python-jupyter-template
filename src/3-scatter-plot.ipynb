### Exercise 3: Scatter Plot with Annotations

```python
# TODO: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

Scatter Plot

```python
import matplotlib.pyplot as plt

# Function to create a scatter plot
def create_scatter_plot():
    try:
        # Data for the scatter plot
        x = [1, 2, 3, 4, 5]
        y = [2, 4, 6, 8, 10]

        # Create the scatter plot
        plt.figure(figsize=(8, 5))
        plt.scatter(x, y, color='green', label='Data Points')

        # Annotate each point
        for i in range(len(x)):
            plt.annotate(f"({x[i]}, {y[i]})", (x[i], y[i]), textcoords="offset points", xytext=(5, 5), ha='center')

        # Customizations
        plt.title("Scatter Plot with Annotations")
        plt.xlabel("X-axis")
        plt.ylabel("Y-axis")
        plt.grid(True)
        plt.legend()

        # Display the plot
        plt.show()

    except Exception as e:
        print(f"An error occurred: {e}")

# Call the function to create the scatter plot
create_scatter_plot()
```
