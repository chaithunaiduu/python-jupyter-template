Subplot

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to create a 2x2 subplot layout
def create_subplots():
    try:
        # Data for each plot
        x = [1, 2, 3, 4, 5]
        y_line = [10, 20, 30, 40, 50]
        categories = ['A', 'B', 'C', 'D']
        values = [15, 25, 35, 45]
        y_scatter = [2, 4, 6, 8, 10]
        labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
        sizes_pie = [40, 25, 20, 15]

        # Create the 2x2 subplot layout
        fig, axs = plt.subplots(2, 2, figsize=(10, 10))

        # Line chart (subplot 1)
        axs[0, 0].plot(x, y_line, marker='o', color='blue', label='y = Line')
        axs[0, 0].set_title("Line Chart")
        axs[0, 0].set_xlabel("X-axis")
        axs[0, 0].set_ylabel("Y-axis")
        axs[0, 0].grid(True)
        axs[0, 0].legend()

        # Bar chart (subplot 2)
        axs[0, 1].bar(categories, values, color='orange', label='Bar Chart')
        axs[0, 1].set_title("Bar Chart")
        axs[0, 1].set_xlabel("Categories")
        axs[0, 1].set_ylabel("Values")
        axs[0, 1].legend()

        # Scatter plot (subplot 3)
        axs[1, 0].scatter(x, y_scatter, color='green', label='Scatter')
        axs[1, 0].set_title("Scatter Plot")
        axs[1, 0].set_xlabel("X-axis")
        axs[1, 0].set_ylabel("Y-axis")
        axs[1, 0].legend()

        # Pie chart (subplot 4)
        axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=140,
                      colors=['gold', 'lightblue', 'lightgreen', 'coral'])
        axs[1, 1].set_title("Pie Chart")

        # Adjust layout
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"An error occurred: {e}")

# Call the function to create the subplots
create_subplots()
```
