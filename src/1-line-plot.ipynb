Line Plot Customization

```python
# Importing required libraries
import matplotlib.pyplot as plt
import numpy as np

# Function to create a line plot
def create_line_plot():
    try:
        # Data for the plot
        x = np.arange(0, 11)  # x values from 0 to 10
        y = x ** 2  # y values as the square of x

        # Create the plot
        plt.figure(figsize=(8, 5))
        plt.plot(x, y, marker='o', color='blue', label='y = x^2')

        # Customizations
        plt.title("Line Plot of y = x^2")
        plt.xlabel("X-axis")
        plt.ylabel("Y-axis")
        plt.grid(True)
        plt.legend()

        # Display the plot
        plt.show()

    except Exception as e:
        print(f"An error occurred: {e}")

# Call the function to create the line plot
create_line_plot()
```
