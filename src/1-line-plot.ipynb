```python
import matplotlib.pyplot as plt
import numpy as np

# Line Plot: x values range from 0 to 10, y values are the square of x
x = np.arange(0, 11)
y = x ** 2

plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', label="y = x^2")
plt.title("Line Plot of y = x^2")
plt.xlabel("x values")
plt.ylabel("y values")
plt.grid(True)
plt.legend()
plt.show()
```
